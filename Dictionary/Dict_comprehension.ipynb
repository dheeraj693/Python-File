## Dictionary comprehension.

```python
## Print 1st 10 number and their square.
d = {i:i*i for i in range(1,11)}
print(d)
```

```python
## Using existing dict.city name and distance in km change in mile.
d = {'Delhi':1000,'Mumbai':2000,'Banglore':3000}
new_d = {key:value*0.62 for (key , value) in d.items() }
print(new_d)
```

```python
## Create  dictionary using two list with the help of zip
days = ['Sunday','Monday','Tuesday','Wednesday','Thursday','Friday','Saturday']
temp = [30,32,34,37,31,40,45]
d = {i:j for (i,j) in zip(days,temp) }
print(d)

        
```

```python
## Next method 
days = ['Sunday','Monday','Tuesday','Wednesday','Thursday','Friday','Saturday']
temp = [30,32,34,37,31,40,45]
d = {}
for key in range(len(days)):
    d[days[key]]=temp[key]
print(d)
```

```python
## Using if condition show products where stock not 0.
product = {'phone':10,'laptop':0,'charger':32,'tablet':0}
new_d = { key:value for (key,value) in product.items() if value>0}
print(new_d)
```

```python
## Print table 
# { 2:{1:2,2:4,3:6,4:8.......}
#   3:{1:3,2:6,3:9,4:12.....}
#  ...................}

n = int(input("Enter the number : "))
d = {
       i:{j:j*i for j in range(1,11)} for i in range(1,n+1)
}
print(d)
```

```python
dict1 = {'a': 10, 'b': 20}
dict2 = {'a': 5, 'c': 15}

merged = dict(dict1)   # copy dict1
for k, v in dict2.items():
    merged[k] = merged.get(k, 0) + v

print(merged)
```

```python
## copy dictionary
d = {0:1,1:2,2:3,3:4}
new = {key:value for key ,value in d.items()}
print(new)
```

```python
d = {0:1,1:2,2:3,3:4}
new = {}
for i,j in d.items():
    new[i]=j
print(new)
```

```python
data = [
("A", 10),
("B", 20),
("A", 15),
("B", 5)
]

d = {}
for x, y in data:
    d[x] = d.get(x, 0) + y  

print(d)
```
