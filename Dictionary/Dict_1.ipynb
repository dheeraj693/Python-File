## Dictionary In Python

```python
## Empty dictionary

d = {}
print(d)
print(type(d))
```

```python
## 1-d Dictionary 
d = {'name':'Dheeraj','gender':'male'}
print(d)
```

```python
d = {(1,2,3):1,"hello":"world"}
print(d)
```

```python
# 2-D dictionary
d = {
    'name':'Dheeraj',
    'collage':'VGU',
    'sem':4,
    'subject':{
        'dsa':50,
        'maths':67,
        'english':43
    }
}
print(d)
```

```python
## Create dictionary using dict()
d = dict([(1,1),(2,2),(3,3)])
print(d)
```

```python
## Duplicate keys are not allowed.
d = {'name':'dheeraj','name':'rahul'}
print(d)
```

#### Access Items not allowed by indexing but access other way.

```python
## Access using key
d = {'name':'Dheeraj','age':22}
print(d['name'])
print(d['age'])
```

#### Adding key value pair

```python
d = {'name':'Dheeraj','age':22}
d['gender'] = 'male'
print(d)
```

```python
d = {'name':'Dheeraj','age':22}
d['gender'] = 'male'
d['weight'] = 66
print(d)
```

#### How can we remove the key value pair

```python
# 1. pop() remove using key 
d = {'name': 'Dheeraj', 'age': 22,3:3, 'gender': 'male', 'weight': 66}
d.pop(3)
print(d)
```

```python
# 2. popitems remove last item
d = {'name': 'Dheeraj', 'age': 22,3:3, 'gender': 'male', 'weight': 66}
d.popitem()
print(d)
```

```python
# 3 . Del and pop work same way.
d = {'name': 'Dheeraj', 'age': 22,3:3, 'gender': 'male', 'weight': 66}
del d['name']
print(d)
```

```python
# clear() used to remove all items
d = {'name': 'Dheeraj', 'age': 22,3:3, 'gender': 'male', 'weight': 66}
d.clear()
print('Dictionary clear : ',d)
```

#### How to eddit item in dictionary.

```python
d = {
    'name':'Dheeraj',
    'collage':'VGU',
    'sem':4,
    'subject':{
        'dsa':50,
        'maths':67,
        'english':43
    }
}
print(d)
print()
d['subject']['dsa']=90
print(d)
```

## Dictionary Operation
#### 1.Membership
#### 2.Iteration

```python
d = {'name': 'Dheeraj', 'age': 22,'id':3, 'gender': 'male', 'weight': 66}
for  i in d:
    print(i,d[i])
```

```python
d = {'name': 'Dheeraj', 'age': 22,'id':3, 'gender': 'male', 'weight': 66}
for  i in d.values():
    print(i)
```

```python
d = {'name': 'Dheeraj', 'age': 22,'id':3, 'gender': 'male', 'weight': 66}
for  i in d.keys():
    print(i)
```

```python
d = {'name': 'Dheeraj', 'age': 22,'id':3, 'gender': 'male', 'weight': 66}
for  i in d.items():
    print(i)
```

```python
d = {'name': 'Dheeraj', 'age': 22,'id':3, 'gender': 'male', 'weight': 66}
for key,values in d.items():
    print(f"{key} = {values}")
```

#### Dictionary Function


```python
# 1 len() = count number of keys
d = {'name': 'Dheeraj', 'age': 22,'id':3, 'gender': 'male', 'weight': 66}
print(len(d))
```

```python
# 2. sorted() 
d = {'name': 'Dheeraj', 'age': 22,'id':3, 'gender': 'male', 'weight': 66}
## Ascending order
print(sorted(d))
```

```python
## Descending order
d = {'name': 'Dheeraj', 'age': 22,'id':3, 'gender': 'male', 'weight': 66}
print(sorted(d,reverse=True))
```

```python
## Key used for show all the key 
d = {'name': 'Dheeraj', 'age': 22,'id':3, 'gender': 'male', 'weight': 66}
d.keys()
```

```python
## Values show all the values
d = {'name': 'Dheeraj', 'age': 22,'id':3, 'gender': 'male', 'weight': 66}
d.values()
```

```python
## .items() used for show the keys and value
d = {'name': 'Dheeraj', 'age': 22,'id':3, 'gender': 'male', 'weight': 66}
d.items()
```

```python
## Create a dictionary for using list
keys = ['name','location']
values = ['Vikash','Jaipur']
d = {}
for key in range(len(keys)):
    d[keys[key]] = values[key]
print(d)
```

```python
##  Swap key and value
d = {'name': 'Vikash', 'location': 'Jaipur'}
new_d = {}
for i,j in d.items():
    new_d[j] = i
print(new_d)
```
