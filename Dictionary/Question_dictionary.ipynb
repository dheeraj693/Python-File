```python
# # Q16. First non-repeating character
# Input: "aabbcde"
s = 'aaabbcde'
for i in range(len(s)):
    repeate = False
    for j in range(len(s)):
        if i!=j and s[i] == s[j]:
            repeate = True
            break
    if not repeate:
        print(s[i])
        break
```

```python
# Q17. Group anagrams1
#Input = ["eat","tea","tan","ate","nat","bat"]
l = ["eat","tea","tan","ate","nat","bat"]

d = {}
for word in l:
    key = ''.join(sorted(word))
    if key in d:
        d[key].append(word)
    else:
        d[key] = [word]
print(d)
#convert into list
print(list(d.values()))
```

```python
## Check Anagram of two string
s1 = 'eat'
s2 = 'aet'
if len(s1) != len(s2):
    print("Not an anagram")
else:
    r1 = ''.join(sorted(s1))
    r2 = ''.join(sorted(s2))
    if r1 == r2:
        print("Anagram")
    else:
        print("Not Anagram")
```
