```python
# 1. First Non-Repeating Character
# Input: "programming" Output: "p"
text = "programming"
d =  {}
for i in text:
    if i in d:
        d[i] += 1
    else:
        d[i] = 1
print(d)
print()
for key,value in d.items():
    if value < 2:
        print("Non  repeatind character : ",key)
        break
```

```python
# 2. First Repeating Character
# Input: "swiss" Output: "s"
text = "swiss"
d =  {}
for i in text:
    if i in d:
        d[i] += 1
    else:
        d[i] = 1
print(d)
print()
for key,value in d.items():
    if value > 2:
        print("Repeatind character : ",key)
        break
```

```python
# 3. Character Frequency
# Input: "apple" Output: {'a':1, 'p':2, 'l':1, 'e':1}
text = 'apple'
d = {}
for i in text:
    if i in d:
        d[i] += 1
    else:
        d[i] = 1
print(d)
```

```python
# 4. Word Frequency in Sentence
# Input: "this is a test this is" Output: {'this':2, 'is':2, 'a':1, 'test':1}

text = "this is a test this is"
word = text.split()
print(word)
print()
d = {}
for i in word:
    if i  in d:
        d[i] += 1
    else:
        d[i] = 1
print(d)
    
```

```python
# 5. Check Anagram
# Input: "listen", "silent" Output: True
text1 = "listen"
text2 = "silent"
if len(text1) != len(text2):
    print("Not Anagram")
else:
    r1 = ''.join(sorted(text1))
    r2 = ''.join(sorted(text2))
    if r1 == r2:
        print(True)
    else:
        print(False)
```

```python
# 6. Maximum Occurring Character
# Input: "banana" Output: 'a'
text = 'banana'
d = {}
for i in text:
    if i in d:
        d[i] += 1
    else:
        d[i] = 1
print(d)
print()
largest = 0
large_key = ''
for key , value in d.items():
    if value > largest:
        largest = value
        large_key = key
print(large_key,largest)
```

```python
# 7. Elements with Frequency > 1
# Input: [1, 2, 2, 3, 4, 4, 4] Output: {2:2, 4:3}
l = [1, 2, 2, 3, 4, 4, 4]
d = {}
for i in l:
     if i in d:
        d[i] += 1
     else:
        d[i] = 1
print(d)
print()
new = {}
for key,value in d.items():
    if value > 1:
        new[key] = value
print(new)
        
        
        
```

```python
# 8. Unique Elements Only
# Input: [1, 2, 2, 3, 4, 4] Output: {1:1, 3:1}
l = [1, 2, 2, 3, 4, 4, 4]
d = {}
for i in l:
     if i in d:
        d[i] += 1
     else:
        d[i] = 1
print(d)
print()
new = {}
for key,value in d.items():
    if value == 1:
        new[key] = value
print(new)
```

```python
# 9. Character Index Mapping
# Input: "hello" Output: {'h':[0], 'e':[1], 'l':[2,3], 'o':[4]}
text = 'hello'
d = {}
for i,j in enumerate(text):
    if j in d:
        d[j].append(i)
    else:
        d[j] = [i]
print(d)
```

```python
# 10. Group Words by Length
# Input: ["cat", "dog", "apple"] Output: {3:['cat','dog'], 5:['apple']}
l = ["cat", "dog", "apple"]
d = {}
for i  in l:
    word = len(i)
    if word in d:
        d[word].append(i)
    else:
         d[word] = [i]
print(d)
```

```python
# 11. Pair Sum Using Dictionary
# Input: [2, 7, 11, 15], target = 9 Output: (2, 7)
l = [2, 7, 11, 15]
target = 9
d = {}
for  i in l:
    num = target - i
    if num in d:
        print("Pair found",(i,num))
        break
    else:
        d[i] = True
```

```python
# 13. Group Anagrams
# Input: ["eat", "tea", "tan", "ate", "nat", "bat"] Output: [['eat','tea','ate'], ['tan','nat'], ['bat']]

l = ["eat", "tea", "tan", "ate", "nat", "bat"]
d = {}
for i in l:
    word = ''.join(sorted(i))
    if word in d:
        d[word].append(i)
    else:
        d[word] = [i]
print(d)
```
