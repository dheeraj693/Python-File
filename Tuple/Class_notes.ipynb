```python
li = [[1,2,3],
      [4,5,6],
      [7,8,9]]
new = []
for row in range(len(li)):
  row1 = []
  for col in range(len(li[row])):
    row1.append(li[col][row])
  new.append(row1)
print(new)
```

```python
matrix = [[1,2,3],
          [4,5,6],
          [7,8,9]]

rows = len(matrix)
cols = len(matrix[0])

# Transpose print
for col in range(cols):          # pehle column fix karo
    new_row = []                 # ek naya row banega transpose ke liye
    for row in range(rows):      # har row ka us column ka element lo
        new_row.append(matrix[row][col])
    print(new_row)               # row print karo
```

```python
matrix = [[1,2,3],
          [4,5,6],
          [7,8,9]]
rows = len(matrix)
cols = len(matrix[0])

for col in range(cols):
    transpose_matrix = []
    for row in range(rows):
        transpose_matrix.append(matrix[row][col])
    print(transpose_matrix)
```
