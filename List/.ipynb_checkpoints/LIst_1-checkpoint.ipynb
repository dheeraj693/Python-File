```python
# 1. Add Elements
# Problem: Add 3 numbers to an empty list using append() .
# Input: 1 2 3 Output: [1, 2, 3]
li = []
for i in range(1,4):
    n = int(input())
    li.append(i)
print(li)
```

```python
# 2. Append vs Extend
# Problem: Add [4,5] to list using:
# append()
# extend()
# # Input: [1,2,3] Output: [1,2,3,[4,5]] (append) [1,2,3,4,5] (extend)
li = []
for i in range(1,4):
    li.append(i)
print(li)
li.append([4,5])
print("Append  : ",li)
li.extend([4,5])
print(li)
```

```python
# 10. Sort Ascending
# Problem: Sort the list.
# Input: [4,1,3,2] Output: [1,2,3,4]
li = [3,1,4,2,5]
print(li)
li.sort()
print(li)
```

```python
## AScending order
li = [3,1,4,2,5]
print(li)
for i in range(len(li)):
    for j in range(0,len(li)-i-1):
        if li[j]>=li[j+1]:
            li[j],li[j+1] = li[j+1],li[j]
print(li)
```

```python
# 11. Sort Descending
# Problem: Sort in descending order using methods.
# Input: [4,1,3,2] Output: [4,3,2,1]
li = [3,1,4,2,5]
print(li)
n = len(li)
for i in range(n):
    for j  in range(0,n-i-1):
        if li[j] < li[j+1]:
            li[j],li[j+1] = li[j+1],li[j]
print(li)
```

```python
# 12. Reverse List
# Problem: Reverse list using reverse() .
# Input: [1,2,3] Output: [3,2,1]
li = [1,2,3,4,5]
print(li)
li.reverse()
print(li)
```

```python
# 13. Extend Two Lists
# Problem: Add all elements of second list into first.
# Input: [1,2] , [3,4] Output: [1,2,3,4]
li = [1,2]
li_2 = [3,4]
li.extend(li_2)
print(li)
```

```python
# 17. Remove All Occurrences
# Problem: Remove all 3 s using methods.
# Input: [3,1,3,2,3] Output: [1,2]
li = [3,1,3,2,3]
for i in li:
    if i==3:
        li.remove(i)
print(li)
```

```python
li = [3,1,3,2,3]
print(li)
new_list = []
for i in li:
    if i!=3:
        new_list.append(i)
print(new_list)
```

```python
# 18. Merge + Sort
# Problem: Merge two lists and sort.
# Input: [3,1] , [4,2] Output: [1,2,3,4]
li_1 = [3,1,2]
li_2 = [4,6,5]
li = li_1 + li_2
print("After merge list 1 and list 2 = ",li)
for i in range(len(li)):
    for j in range(0,len(li)-i-1):
        if li[j] > li[j+1]:
            li[j],li[j+1] = li[j+1],li[j]
print("After sorting =",li)
```

```python
# 19. Check Element Exists
# Problem: Check if 5 exists using count() .
# Input: [1,2,3,4] Output: False
li = [1,2,3,4]
print(li)
count = False
for i in li:
    if i == 5:
        count = True
print(count)
```

```python
# 20. Replace Element
# Problem: Replace first occurrence of 2 with 99 .
# Input: [1,2,3,2] Output: [1,99,3,2]
li = [1,2,3,2]
for i in li:
    if li[i]==2:
        li[i] = 99
        break
print(li)
```
