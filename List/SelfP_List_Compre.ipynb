```python
## Q1. Even numbers nikaalo list se
nums = [1,2,3,4,5,6]
even = [x for x in nums if x%2 == 0]
print(even)
```

```python
## Q2. Odd numbers ka square banao
nums = [1,2,3,4,5,6]
square_odd = [x*x for x in nums if x%2 != 0]
print(square_odd)
```

```python
## Q3. Words me se sirf length > 3 wale lo
words = ["hi", "hello", "cat", "python"]
w = [word for word in words if len(word) > 3]
print(w)
```

```python
## Q4. Numbers jo 5 se bade hain unko lo
nums = [2,7,3,9,1,20,11,23]
num = [ x for x in nums if x > 5]
print(num)
```

```python
## Q5. Even → same, Odd → 0
nums = [2,7,3,9,1,20,11,23]
even = [i if i%2==0 else -1 for i  in nums]
print(even)
```

```python
## Q6. Positive → same, Negative → "neg"
nums = [1,-2,3,-4,-5,5,6,-6]
even = [i if i > 0 else 'neg' for i in nums]
print(even)
```

```python
## Q7. String me vowels ko "V" aur consonant ko "C" se replace karo
words = ["hi", "hello", "cat", "python"]
vowels = 'aeiou'
new = [''.join(['V' if j in vowels else 'C'  for j in i]) for i in words]
print(new)
```

```python
## Q10. Negative numbers ko 0 karo, positive same rakho
nums = [1,-2,3,-4,5,-6,7,8]
new = [0 if i<0 else i for i in nums]
print(new)
```

```python
## Q11. 2D list flatten karo
l = [[1,2,3],[4,5,6],[7,8,9]]
new = [l[i][j] for i in range(len(l)) for j in range(len(l[i]))]
print(new)
```

```python
## Q12. Har word ke vowels extract karo
words = ["hi", "hello", "cat", "python"]
vowels = 'aeiou'
new = [''.join([j for j in x if j in vowels])for x in words]
print(new)
```

```python
## Q14. Numbers divisible by 3 and 5
num = int(input())
new = [n for n in range(1,num+1) if n%3==0 and n%5==0]
print(new)
```

```python
## Q15. Even numbers ko rakho, odd ko remove mat karo — unko -1 bana do
n = int(input())
new = [i if i%2==0 else -1 for i in range(1,n+1)]
print(new)
```
