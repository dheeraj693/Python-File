```python
# Check a element in a list is palindrome or not
l = [1,2,3,4,2,2,1]
n = len(l)
mid = n//2
left = []
right = []
print(mid)
for i in range(n):
    left.append(l[i])
    if i == mid:
        break
for j in range(n-1,mid-1,-1):
    right.append(l[j])
    
print(left)
print(right)
if left == right:
    print("Palindrome of a list")
else:
    print("Not a palindrome of a list")
    
```

```python
## Count frequency of each element
l = [1,2,3,4,5,4,2,1,1,4]
n = len(l)
for i in range(n):
    for j in range(0,n-i-1):
        if l[j] > l[j+1]:
            l[j],l[j+1] = l[j+1],l[j]
sort_l = l
print("Sorted List = ",sort_l)
count = 1
counter = []
for i in range(len(sort_l)-1):
    if sort_l[i] == sort_l[i+1]:
        count += 1
        
    else:
        counter.append(count)
        count = 1
counter.append(count)
print(counter)
        
```

```python
## Frequency count each alement of a list using dictionary
l = [1,2,3,4,5,4,2,1,1,4]
freq_dict = {}

for i in range(len(l)):
    if l[i] in freq_dict:
        freq_dict[l[i]] +=  1
    else:
        freq_dict[l[i]] = 1

print(freq_dict)
    
```

```python
# Q6. Find second largest element
l = [4,10,3,15,11,50]
largest = 0
sec_largest = 0
for i in range(len(l)):
    if l[i] > largest:
        sec_largest = largest
        largest = l[i]
        
    elif l[i]>largest and l[i]  > sec_largest:
        sec_largest = l[i]
print(sec_largest)
        
```

```python
# Q7. Move all zeros to end

li = [0,1,0,3,12]

n = len(li)
revs = []

for i in range(len(li)):
    if li[i] > 0:
        revs.append(li[i])

zero = li.count(0)
while zero > 0:
    revs.append(0)
    zero -= 1
print(revs)

```
