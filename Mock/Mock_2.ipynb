```python
# Q1. Hollow Diamond Pattern (8 Marks)
# Write a Python program to print a hollow diamond pattern for a given odd integer n

n = 7
# First Half 
for i in range(1,n+1):
    for j in range(n-i):
        print(' ',end=' ')
    for j in range(2*i-1):
        if j == 0 or j == 2*i-2:   # first or last position
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
# Second Half
for i in range(n-1,0,-1):
    for j in range(n-i):
        print(' ',end=' ')
    for j in range(2*i-1):
        if j == 0 or j == 2*i-2:   # first or last position
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
 
```

```python
# Q2. String Manipulation – Longest Unique Word (6 Marks)
# Input: "hello world, python programming is fun"
# ● Output: "world"
text = "hello world  programming is fun"

words = text.split()

longest_word = ""

for word in words:
    if len(word) == len(set(word)):
        if len(word) > len(longest_word):
            longest_word = word

print(longest_word)
        
    
```

```python
# 1. Use filter() to keep only prime numbers.
def prime(num):
    is_prime = True
    if num < 2:
        return False
    for j in range(2, num):
        if num % j == 0:
            is_prime = False
            break
    return is_prime  

li = [2,3,4,5,6,7,8]
list(filter(prime , li))
```

```python
# ● A function that returns the factorial of each number.
def fact(n):
     if n==0 or n==1:
         return 1
     else:
         return n*fact(n-1)

    
def filter(func , li):
    result = []
    for i in li:
        result.append(func(i))
    return result

li = [1,2,3,4,5]
filter(fact , li)
```

```python
# Q5. Simple Decorator – Welcome Message (8 Marks)
# Create a decorator greet_decorator that:
# 1. Prints "Function is starting..." before executing the function.
# 2. Prints "Function execution completed." after executing the function.
def decorator(func):
    def wraper():
        print("Function is starting...")
        func()
        print("Function execution completed.")
    return wraper

@decorator
def say_hello():
    print("Hello Dheeraj")
say_hello()
```

```python
# Q7. Tuple + Sorting – Student Ranking (6 Marks)
# Given a list of tuples in the form:
# [("Vinay", 85), ("Amit", 92), ("Ravi", 78), ("Neha", 92)]

# Sort the data:
# 1. By marks in descending order.
# 2. If marks are equal, by name alphabetically.
# Return the sorted result.
```

```python
l = [("Vinay", 85), ("Amit", 92), ("Ravi", 78), ("Neha", 92)]

# List to dictionary
d = dict(l)
items = list(d.items())

n = len(items)
for i in range(n):
    for j in range(i+1, n):
        # Sort by value
        if items[i][1] < items[j][1]:
            items[i], items[j] = items[j], items[i]
        # Sort by name
        elif items[i][1] == items[j][1] and items[i][0] > items[j][0]:
            items[i], items[j] = items[j], items[i]

# Print sorted result
for name, marks in items:
    print(name, marks)
```

```python
# Q8. Balanced Parentheses (6 Marks)
# Write a function to check whether a string containing only ( and ) is balanced.
# Example:
# ● Input: "(()())"
# ● Output: True
# ● Input: "(()"
# ● Output: False

def check_braces(s):
    
    count = 0
    for ch in s:
        if ch == "(":
            count += 1
        elif ch == ")":
            count -= 1
            if count < 0:  
                return False
    return count == 0
print(check_braces('(()())'))
    
```

```python
# Q9. List + Dictionary – Group Anagrams (4 Marks)
# Group all anagrams together.
# Example:
# ["eat", "tea", "tan", "ate", "nat", "bat"]
l= ["eat", "tea", "tan", "ate", "nat", "bat"]
d = {}
for i in l:
    word = ''.join(sorted(i))
    if word in d:
        d[word].append(i)
    else:
        d[word] = [i]
print(d)
print()
result = []
for key , value in d.items():
    result.append(value)
print(result)
```

```python
# Q10. Functional Programming Challenge (4 Marks)
# Given a list of strings:
# 1. Use filter() to keep only palindromes.
# 2. Use map() to convert them to uppercase.
# 3. Return the final list sorted by length.
# Example:
# ["madam", "python", "level", "code", "radar"]

# Output:
# ['LEVEL', 'MADAM', 'RADAR']

def palindrome(num):
    return num == num[::-1]

l = ["madam", "python", "level", "code", "radar"]

result = list(filter(palindrome, l))
print(result)

new = list(map(str.upper, result))
print(new)

# new = []
# for i in result:
#     new.append(i.upper())

#print(new)
```
