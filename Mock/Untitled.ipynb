```python
n = int(input("Enter the number :"))
m = int(len(str(n)))
arm = 0
temp = n
while n>0:
    digit = n%10
    arm = arm + digit**m
    n = n//10
if temp == arm:
    print("Armstrong number")
else:
    print("Not armstrong number")
```

```python
price = int(input("Enter the price : "))
if price > 2000:
    print("20 percent discount")
    new_price = ((100-20)*price)/100
    print("New Price = ",new_price)
elif price > 1000:
    print("10 percent discount")
    new_price = ((100-10)*price)/100
    print("New Price = ",new_price)
elif price > 500:
    print("5 percent discount")
    new_price = ((100-5)*price)/100
    print("New Price = ",new_price)
else:
    print("No discount")
    
```
