```python
# Q1. Hollow Diamond Pattern (8 Marks)
# Write a Python program to print a hollow diamond pattern for a given odd integer n

n = 7
# First Half 
for i in range(1,n+1):
    for j in range(n-i):
        print(' ',end=' ')
    for j in range(2*i-1):
        if j == 0 or j == 2*i-2:   # first or last position
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
# Second Half
for i in range(n-1,0,-1):
    for j in range(n-i):
        print(' ',end=' ')
    for j in range(2*i-1):
        if j == 0 or j == 2*i-2:   # first or last position
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
 
```

```python
# Q2. String Manipulation – Longest Unique Word (6 Marks)
# Input: "hello world, python programming is fun"
# ● Output: "world"
text = "hello world  programming is fun"

words = text.split()

longest_word = ""

for word in words:
    if len(word) == len(set(word)):
        if len(word) > len(longest_word):
            longest_word = word

print(longest_word)
        
    
```

```python
# Q3. List Processing with map() and filter() (6 Marks)
# Given a list of integers, perform the following:
# 1. Use filter() to keep only prime numbers.
# 2. Use map() to square each prime number.
# 3. Return the resulting list.
# Example:
# ● Input: [2, 3, 4, 5, 6, 7]
# ● Output: [4, 9, 25, 49]

#2. Use map() to square each prime number.
def square(num):

    return num**2
    
li = [2,3,4,5,6,7]
list(map(square , li))
```

```python
# 1. Use filter() to keep only prime numbers.
def prime(num):
    result = []
    is_prime = True
    for i in range(2,num):
        for j in range(2,i):
            if i%j == 0:
                is_prime = False
                break
        if is_prime:
            result.append(i)

li = [2,3,4,5,6,7,8]
list(filter(prime , li))
```
