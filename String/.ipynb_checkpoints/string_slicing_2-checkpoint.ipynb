```python
#33. Swap first and last character using slicing.
s = 'PYTHONPROGRAMMING'
print(len(s))
print(s[0])
print(s[16])
l = list(s)
print(l)
l[0],l[16] = l[16],l[0]
s = ''.join(l)
print(s)
```

```python
#34. Swap first two and last two characters using slicing.
s = 'PYTHONPROGRAMMING'
print(len(s))
print(s[0])
print(s[16])
l = list(s)
print(l)
l[1],l[15] = l[15],l[1]
l[0],l[16] = l[16],l[0]
#l[1],l[15] = l[15],l[1]
s = ''.join(l)
print(s)
```
