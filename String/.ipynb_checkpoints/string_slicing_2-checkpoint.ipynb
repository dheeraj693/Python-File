```python
#11. Reverse the complete string using slicing.
s = 'PYTHONPROGRAMMING'
print(s[::-1])
```

```python
#12. Reverse the first 4 characters of a string.
s = 'PYTHONPROGRAMMING'
print(s[-14:-18:-1])

#OR

print()
print(s[0:4][::-1])

```

```python
#13. Reverse the last 5 characters of a string.
s = 'PYTHONPROGRAMMING'
print(s[:-6:-1])


#OR
print()
print(s[len(s):-6:-1])
```

```python
#14. Reverse the string except the first character.
s = 'PYTHONPROGRAMMING'
print(s[-1:-17:-1])

# OR
print()
print(s[1:len(s)][::-1])
```

```python
#15. Reverse the string except the last character.
s = 'PYTHONPROGRAMMING'
print(s[-2:-18:-1])
#OR
print()

print(s[0:len(s)-1][::-1])
```
