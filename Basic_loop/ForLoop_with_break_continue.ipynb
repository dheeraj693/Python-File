# Break and Continue Coding Questions

```python
# 1. Print numbers from 1 to 10, but stop the loop when the number becomes 6.

for  i in range(1,11):
    if i == 6:
        break
    print(i)
```

```python
# 2. Print numbers from 1 to 10, but skip printing 5.

for  i in range(1,11):
    if i == 5:
        continue
    print(i)
```

```python
# 3. Print numbers from 1 to 20, but stop when a number is divisible by 7.
n = int(input("Enter the number : "))
for i in range(1,n+1):
    if i % 7 == 0:
        break
    print(i)
```

```python
# 4. Print numbers from 1 to 20, skipping all even numbers.
n = int(input("Enter the number : "))
for i in range(1,n+1):
    if i % 2 == 0:
        continue
    print(i)
```

```python
# 11. Print numbers from 1 to 30, but stop when the number is a perfect square.
n = int(input("Enter the number : "))
for i in range(2,n+1):
    perfect_square = False
    for j in range(1,i+1):
        if j*j == i:
            perfect_square = True
            break
    if perfect_square == True:
        break
    print(i)
            
```

```python
# Another method for question 11. by square root method.

n = int(input("Enter the number : "))
for i in range(2,n+1):
    square_root = int(i**0.5)
    if square_root * square_root == i:
        break
    print(i)
```

```python
# 15. Print numbers from 1 to 50, but stop when the number contains the digit 9.
n = int(input("Enter the number : "))
for i in range(10,n+1):
    temp = i
    while n > 0:
        digit = temp%10
        if digit ==  9:
            break
        temp = temp//10
    if digit == 9:
        break
    print(i)
```

```python
# 16. Print numbers from 1 to 20, skipping numbers whose sum of digits is even.
n = int(input("Enter the number : "))

for i in range(1,n+1):
    temp = i
    sum = 0
    while temp > 0:
        digit = temp%10
        sum += digit
        temp  = temp // 10
    if sum % 2 == 0:
        continue
    print(i)
```

```python
# 17. Print numbers from 1 to 100, but stop when a number is divisible by both 8 and 12.
n = int(input("Enter the number : "))
for i in range(1,n+1):
    if i % 8 == 0 and i % 12 == 0:
        break
    print(i ,end=' ')
```

```python
# 18. Print numbers from 1 to 50, skipping numbers that are perfect squares.
n = int(input("Enter the number : "))
for i in range(1,n+1):
    square_root = int(i**0.5)
    if i == square_root*square_root:
        continue
    print(i , end=' ')
```

```python
# 19. Print numbers from 1 to 100, but stop when the number is a palindrome.
n = int(input("Enter the number : "))
for i in range(1,n+1):
    temp = i
    reverse = 0
    while temp > 0:
        digit = temp%10
        reverse = reverse * 10 + digit
        temp = temp//10
    if i == reverse:
        continue
    print(i)
        
```

```python
# 20.Print numbers from 1 to 30, skipping numbers whose factorial is greater than 100.
n = int(input("Enter the number : "))
for i in range(1,n+1):
    fact = 1
    for j in range(1,i+1):
        fact *= j
    if fact > 100:
        continue
    print(i)
```
