## Pattern printing in pyramid form 

```python
# 1.
for i in range(1,5):
    for j in range(4-i):
        print(' ',end=' ')
    for k in range(2*i-1):
        print(i,end=' ')
    print()
```

```python
# 2.
for i in range(1,5):
    for j in range(4-i):
        print(' ',end=' ')
    for k in range(2*i-1):
        print(5-i,end=' ')
    print()
```

```python
# 3.
for i in range(1,5):
    for j in range(4-i):
        print(' ',end=' ')
    for k in range(2*i-1):
        print(chr(64+i),end=' ')
    print()
```

```python
# 4.
for i in range(1,5):
    for j in range(4-i):
        print(' ',end=' ')
    for k in range(1,i+1):
        print(k,end=' ')
    for k in range(i-1,0,-1):
        print(k,end=' ')
    print()
```

```python
# 5.
for i in range(1,5):
    for j in range(4-i):
        print(' ',end=' ')
    for k in range(1,i+1):
        print(chr(64+k),end=' ')
    for k in range(i-1,0,-1):
        print(chr(64+k),end=' ')
    print()
```

```python
# 5.
for i in range(1,5):
    for j in range(4-i):
        print(' ',end=' ')
    for k in range(2*i-1):
        if k%2 == 0:
            print(1,end=' ')
        else:
            print(0,end=' ')
    print()
```
