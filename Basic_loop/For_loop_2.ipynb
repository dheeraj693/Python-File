### Basic for Loop and range() Usage

```python
#1. Write a program to print numbers from 1 to 10 using a for loop.
for i in range(1,11):
    print(i,end=' ')
```

```python
#2. Print numbers from 0 to 5 using range().
for i in range(0,6):
    print(i,end=' ')
```

```python
#3. Print numbers from 5 to 15.
for i in range(5,16):
    print(i,end = ' ')
```

```python
#4. Print numbers from 10 to 1 using a for loop.
for i  in range(10,0,-1):
    print(i,end =' ')
```

```python
#5. Print all even numbers between 1 and 20.
for i in range(1,21):
    if i %2 ==0:
        print(i,end = ' ')
```

```python
#6. Print all odd numbers between 1 and 20.
for i in range(1,21):
    if i%2==1:
        print(i,end=' ')
```

```python
#7. Print numbers from 1 to 50, skipping multiples of 5.
for i in range(1,51):
    if i%5==0:
        continue
    print(i,end=' ')    
```

```python
#8. Print numbers from 1 to 100 with a step of 10.
for i in range(1,101,10):
    print(i,end=' ')
```

```python
#9. Print the square of numbers from 1 to 10.
for i  in range(1,11):
    print(f"Square of {i} = {i*i}")
```

```python
# 10. Print the cube of numbers from 1 to 5.
for i  in range(1,11):
    print(f"Cube of {i} = {i*i*i}")
```

### Aggregation and Counting

```python
#11. Find the sum of numbers from 1 to 10 using a for loop.
sum_num = 0
for i in range(1,11):
    sum_num += i
print(f"Sum of 1 to 10 = {sum_num}")
```

```python
#12. Find the sum of even numbers from 1 to 20.
even_sum = 0
for i in range(1,20):
    if i%2==0:
        even_sum += i
print("Sum of even number",even_sum)
```

```python
#13. Print the multiplication table of 5.
n = int(input("enetr number "))
print(f"You enter {n}")
for i in range(1,11):
    print(f'{n} X {i} = {n*i}')
```

```python
#14. Print multiplication tables from 1 to 3.
n = int(input("enetr number "))
print(f"You enter {n}")
for i in range(1,n+1):
    for j in range(1,11):
        print(f'{i} X {j} = {i*j}')
    print("--------------------------------------------")
```

```python
#15. Count how many numbers are there between 1 and 50.
count = 0
for i in range(1,51):
    count += 1
print(f"Total number from 1 to 50 = {count}")
```

```python
#16. Count how many even numbers are there between 1 and 100.
count = 0
for i in range(1,101):
    if i%2==0:
        count += 1
print(f"Total even number from 1 to 100 = {count}")
```

### Control Flow and Formatting

```python
# 17. Print numbers from 1 to 30, but stop printing when the number becomes 20.
for i in range(1,31):
    print(i,end=' ')
    if i == 20:
        break
```

```python
# 18. Print numbers from 1 to 20, but do not print 10.
for i in range(1,21):
    if i == 10:
        continue
    print(i,end=' ')
```
