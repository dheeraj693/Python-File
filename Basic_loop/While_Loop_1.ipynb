## WHILE LOOP – BASIC QUESTIONS (25 Questions)

```python
# 1. Write a program to print numbers from 1 to 10 using a while loop.
n = int(input("Enter the number :  "))
i = 1
while i <= n:
    print(i,end  =' ')
    i += 1
```

```python
#11. Find the sum of even numbers from 1 to 20 using a while loop.
n = int(input("Enter the number :  "))
i = 1
sum_even = 0
while i <= n:
    if i % 2 == 0:
        sum_even += i
        
    i += 1
print("Sum of even number :",sum_even)
```

```python
# 13. Print multiplication tables from 1 to 3 using a while loop.
n = int(input("Enter the number :  "))
i = 1
while i <= n:
    j = 1
    while j <= 10:
        print(f"{i} x {j} = {i*j}")
        j += 1
    
    i += 1
```

```python
#17. Print numbers from 1 to 20, but skip printing 10.
n = int(input("Enter the number :"))
i = 1
while i <= n:
    if i == 10:
        i += 1
        continue
    print(i,end=' ')
    i += 1
```
