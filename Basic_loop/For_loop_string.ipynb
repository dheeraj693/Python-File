## BASIC LEVEL

```python
# 1. Print all vowels from a given string using a for loop.
s = "Hello Python"
vowels = 'aeiou'
c_vowels = ' '
for i in s:
    if i in vowels:
        c_vowels += i
print("All vowels are : ",c_vowels)
```

```python
# 2. Count total number of vowels in a string.
s = "Hello Python"
vowels = 'aeiou'
c_vowels = 0
for i in s:
    if i in vowels:
        c_vowels += 1
print("All consonant are : ",c_vowels)
```

```python
# 3. Count total number of consonants in a string.
s = "Hello Python"
vowels = 'aeiou'
c_consonant = 0
for i in s:
    if i not in vowels:
        c_consonant += 1
print("Number of All consonant are : ",c_consonant)
```

```python
# 4. Print each character of a string, but skip vowels using continue.
s = "Hello Python"
vowels = 'aeiou'

for i in s:
    if i in vowels:
        continue
    print(i , end=' ')
```

```python
# 5. Print only vowels and skip consonants using continue.
s = "Hello Python"
vowels = 'aeiou'
for i in s:
    if i not in vowels:
        continue
    print(i,end=' ')
```

```python
# 6. Print characters until the first vowel is found using break.
s = "Hello Python"
vowels = 'aeiou'
for i in s:
    if i in vowels:
        break
    print('First vowels found  :',i)
        
        
```

```python
# 7. Print characters until the first consonant is found using break.
s = "Hello Python"
vowels = 'aeiou'
for i in s:
    if i not in vowels:
        break
print('First consonant found  :',i)
```

```python
# 8. Count vowels and consonants separately using a loop.
s = input("Enter the string")
vowels = 'aeiou'
c_vowel = ''
c_consonant = ''

for i in s:
    if i in vowels.lower():
        c_vowel += i
    else:
        c_consonant += i
print('All vowels are  :',c_vowel)
print('All consonant are :',c_consonant)
```

## MEDIUM LEVEL

```python
# 9. Print all vowels in uppercase and consonants in lowercase.
s = input("Enter the string")
vowels = 'aeiou'
c_vowel = ''
c_consonant = ''

for i in s:
    if i in vowels.lower():
        c_vowel += i
    else:
        c_consonant += i
print('All vowels are  :',c_vowel.upper())
print('All consonant are :',c_consonant.lower())
```

```python
# 10. Stop printing the string when a vowel appears using break.
s = input("Enter the string :")
vowels = 'aeiou'

for i in s:
    if i in vowels:
        break
    print(i ,end ='' )
```

```python
# 11. Skip digits and special characters, and count only vowels and consonants.
s = input("Enter the string :")
vowels = 'aeiou'
count_v =0
count_c =0
for i in s:
    if i.isnumeric():
        continue
    if not i.isalpha():
        continue
    if i in vowels.lower():
        count_v += 1
    else:
        count_c += 1
print("All vowels are :",count_v)
print('All consonant are :',count_c)
```

```python
# 12. Print only first 3 vowels from a string using break.
s = input("Enter the string :")
vowels = 'aeiou'
c_vowels = 0
for i in s:
    if i.lower() in vowels:
        print(i,end='')
        c_vowels += 1
    if c_vowels == 3:
        break


```

```python
# 13. Skip repeated vowels using continue.
s = input("Enter the string :")
vowels = 'aeiou'
c_vowels = ''
for i in s.lower():
    if i in vowels:
        if i in c_vowels:
            continue
        print(i,end='')
        c_vowels += i
```

```python
# 14. Print characters except the first vowel encountered.
s = input("Enter the string :")
vowels = 'aeiou'
first_vowel = True
for i in s:
    if i in vowels.lower() and first_vowel:
        first_vowel = False
        continue
    print(i,end='')
    
```

```python
# 15. Find the first vowel and stop the loop, print its index.
s = input("Enter the string :")
vowels = 'aeiou'

for i in range(len(s)):
    if s[i] in vowels:
        print(f'First vowels {s[i]} found at index {i}')
        break
        
```

```python
# 16. Print all consonants except the first two consonants (use continue).
s = input("Enter the string :")
vowels = 'aeiou'
count_c = 0
for i in s:
    if i not in vowels.lower():
        if count_c < 2:
            count_c += 1
            continue
        print(i,end='')
    
```

```python
# 17. Reverse a string but skip vowels using continue.
s = input("Enter the string :")
vowels = 'aeiou'
for i in range(len(s)-1,-1,-1):
    if s[i] in vowels:
        continue
    print(s[i],end = '')
    
    
    
```

## INTERVIEW LEVEL

```python
# 18. Check if a string contains at least one vowel, stop as soon as found (break).
s = input("Enter the string :")
vowels = 'aeiou'
for i in s:
    if i.lower() in vowels:
        break
    print(i,end='')
```

```python
# 19. Count vowels until a digit appears, then stop (break).
s = input("Enter the string :")
vowels = 'aeiou'
count_v = ''
count = 0
for i in s:
    if i.isnumeric():
        break
    if i in vowels:
        count_v += i
        count += 1
print("Vowels are :",count_v)
print("total vowels :",count,end='')
```

```python
#  20. Print positions of vowels but skip if vowel repeats consecutively (continue).
s = input("Enter the string :")
vowels = 'aeiou'
for i in range(len(s)):
    if s[i].lower() in vowels:
        if i+1 < len(s) and s[i].lower() == s[i+1].lower():
            continue
        print(s[i],i)
```

```python
# 21. Replace vowels with * but skip if vowel is 'a' using continue. 
s = input("Enter the string :")
vowels = 'aeiou'
new_string = ''
for i in s.lower():
    if i in vowels:
        if i == 'a':
            continue
        new_string += "*"
    else:
        new_string += i
print(new_string)
```

```python
#22. Count consonants but ignore uppercase letters using continue.
s = input("Enter the string :")
vowels = 'aeiou'
count = 0
for i in s:
    if i.lower() in vowels:
        continue
    elif i in s.upper():
        continue
    else:
        count += 1
print(count)
```

```python
#23. Print string until 2 vowels are found, then break.  
s = input("Enter the string :")
vowels = 'aeiou'
count = 0
for i in s:
    if i in vowels:
        count += 1
        if count==2:
            break
    print(i ,end = '')
```

```python
#  24. Skip vowels at even index positions using continue.
s = input("Enter the string :")
vowels = 'aeiou' 
for i in range(len(s)):
    if s[i] in vowels:
        if i % 2 == 0 :
            continue
    print(s[i],i)
```
