```python
# reverse the every word from  the given seenetnce
sentence = 'cat sat on mat'
every_word = sentence.split()
print(every_word)
rev_word = []
for word in every_word:
    rev = ''
    for j in range(len(word)-1,-1,-1):
        rev += word[j]
    rev_word.append(rev)
result = ' '.join(rev_word)
print(result)
        
        
```

```python
sentence = 'cat sat on mat'
print(sentence)
every_word = sentence.split()
rev_word = []
for word in every_word:
    rev_word.append(word[::-1])
rev_sent = ' '.join(rev_word)
print(rev_sent)
```

```python
# Print the largest word from the list
li = ['apple','elephant','cat']
largest = li[0]
for x in li:
    if len(x) > len(largest):
        largest = x
print("Largest word : ",largest)
    
```
