## While Loop Practice Questions

```python
#1. Print numbers from 1 to 100 that are divisible by 3 using a while loop.
n = int(input("Enter the number : "))
i = 1
while i <= n:
    if i % 3 == 0:
        print(i,end=' ')
    i += 1
```

```python
#2. Print numbers from 1 to 50 that are divisible by 5 but not by 10 using a while loop.

n = int(input("Enter the number : "))
i = 1
while i <= n:
    if i % 5 == 0:
        print(i,end=' ')
    i += 1
```

```python
#3. Find the sum of digits of a given number using a while loop.
n = 12345
sum_digit =  0
while n > 0:
    digit = n%10
    sum_digit = sum_digit+digit
    n = n//10
    i += 1
print("Sum of digit :",sum_digit)
```

```python
#4. Reverse a given number using a while loop.
n = 12345
m = n
reverse = 0
while m>0:
    digit = m%10
    reverse = reverse*10 + digit
    m = m//10
    i += 1
print(f"Reverse of the {n} is : {reverse}")
```
