```python
#5. Check whether a given number is a palindrome using a while loop.
n = int(input("Enter a number: ")) 
m = n
reverse = 0
i = 1

while n > 0:
    digit = n % 10
    reverse = reverse * 10 + digit
    n = n // 10
    i += 1
if reverse == m:
    print(f"Number is polindrome : {reverse}")
else:
    print(f'Number is Not a polindrome : {reverse}')
```

```python
#13. Check whether a given number is a perfect number using a while loop.
# Perfect number means sum of factor is equal to the original number 
# 6 factor(1 , 2 , 3) => 1+2+3 = 6

n = int(input("Enter the number : "))
i = 1
factor_sum = 0
while i<n:
    if n % i  ==  0:
        factor_sum += i
    i += 1
print(f'Factor sum of {n} = {factor_sum}')
if factor_sum == n:
    print(f"{n} is perfect number")
else:
    print(f"{n} is not a perfect number")
```

```python
#14. Print numbers from 1 to 100 that are perfect squares using a while loop.
n = int(input("Enter the number : "))
i = 1
while i<n:
    square_root = int(i**0.5)
    if square_root * square_root == i:
        print(i , end = ' ')
    i += 1
```
