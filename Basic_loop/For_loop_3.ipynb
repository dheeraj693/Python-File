```python
#9. Print numbers from 1 to 100 that are palindromes (single loop).
for i in range(1,101):
    reverse = 0
    num = i
    while num>0:
        digit = num%10
        reverse = reverse*10 + digit
        num = num//10
    if reverse == i:
        print(i,end=' ')
```

```python
#24.Print numbers from 1 to 100 that are triangular numbers.

i = 1
triangular = 0

print("Triangular numbers from 1 to 100 are:")
while True:
    triangular = i * (i + 1) // 2   # formula
    if triangular > 100:
        break
    print(triangular)
    i += 1
```

```python
print("Triangular numbers from 1 to 100 are:")

for i in range(1, 100):   # loop 1 se start
    triangular = i * (i + 1) // 2   # formula Tn = n(n+1)/2
    if triangular > 100:   # agar 100 se bada ho gaya toh loop break
        break
    print(triangular)
```

```python
# Check prime or not using finding factor methods
n = int(input("Enter the number : "))
factor = 0
for i in range(1,n+1):
    if n % i == 0:
        factor += 1
if factor == 2:
    print("Prime number")
else:
    print("Not a prime number")
```

```python
# Check all prime number between 1 to n.
n = int(input("Enter the number :"))
for i in range(2,n+1):
    is_prime = True
    for j in range(2,i):
        if i%j==0:
            is_prime = False
            break
    else:
        print(i,end=' ')
```
