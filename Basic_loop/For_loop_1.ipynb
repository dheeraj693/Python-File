## FOR LOOP BASIC QUESTIONS (20–25)

### Basic for Loop and range() Usage

```python
#1. Write a program to print numbers from 1 to 10 using a for loop.
for  i  in range(1,11):
    print(i)    
```

```python
#2. Print numbers from 0 to 5 using range().
for i in range(6):
    print(i)
    
```

```python
#3. Print numbers from 5 to 15.
for i in range(5,16):
    print(i)
```

```python
#4. Print numbers from 10 to 1 using a for loop.
for i in range(10,0,-1):
    print(i)
```

```python
#5. Print all even numbers between 1 and 20.
for i in range(1,21):
    if i % 2 == 0:
        print(i)    
```

```python
#6. Print all odd numbers between 1 and 20.
for i in range(1,21):
    if i % 2 != 0:
        print(i)    
```

```python
#7. Print numbers from 1 to 50, skipping multiples of 5.
for i in range(1,51):
    if i % 5 == 0:
        continue
    print(i , end=' ')
```

```python
# Program to print numbers from 1 to 100 with a steps 10
for i in range(1,101,10):
    print(i)
```

```python
#9. Print the square of numbers from 1 to 10.
for i in range(1,11):
    print(i**2)
```

```python
#10. Print the cube of numbers from 1 to 5.
for i in range(1,6):
    print(i**3)
```

###  Aggregation and Counting

```python
#11. Find the sum of numbers from 1 to 10 using a for loop.
sum = 0
for i in range(1,11):
    sum += i
print("The sum of numbers from 1 to 10 is:", sum)
```

```python
#12. Find the sum of even numbers from 1 to 20.
sum_even = 0
for i in range(1,21):
    if i % 2 == 0:
        sum_even += i
print("The sum of even numbers from 1 to 20 is:", sum_even)
    
```

```python
#13. Print the multiplication table of 5.
n = 5
for i in range(1,11):
    print(f"{n} X {i} = {n*i}")
```

```python
#14. Print multiplication tables from 1 to 3.
for n in range(1,4):
    print(f"Multiplication Table of {n}:")
    for i in range(1,11):
        print(f"{n} X {i} = {n*i}")
    print()  
```

```python
#15. Count how many numbers are there between 1 and 50.
count = 0
for i in range(1,51):
    count += 1
print("The count of numbers between 1 and 50 is:", count)
```

```python
#16. Count how many even numbers are there between 1 and 100.
even_count = 0
for i in range(1,101):
    if i % 2 == 0:
        even_count += 1
print("The count of even numbers between 1 and 100 is:", even_count)
```

### Control Flow and Formatting

```python
# 17. Print numbers from 1 to 30, but stop printing when the number becomes 20.
for i in range(1,31):
    if i > 20:
        break
    print(i)
```

```python
#  18. Print numbers from 1 to 20, but do not print 10.
for i in range(1,21):
    if i == 10:
        continue
    print(i)
```

```python
# 19. Print numbers from 1 to 5, each on a new line.
for i in range(1,6):
    print(i)
```

```python
# 20.Print numbers from 1 to 5 in a single line separated by space.
for i in range(1,6):
    print(i , end='  ')
```

```python
# 21. Print numbers from 5 to 1 in reverse order.
for i in range(5,0,-1):
    print(i,end = '  ') 
```

```python
# 22.Print numbers from 1 to 5, along with their squares.

for i in range(1,6):
    print(f"{i} and its square is {i**2}")
```
