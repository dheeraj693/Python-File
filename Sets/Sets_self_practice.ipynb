## Sets In Python.

```python
## Create a set
s = {1,2,3,4,5}
print(s)

print()

s = set([1,2,3,4,5])
print(s)
```

```python
## Creat empty set by using set constructor.....
s = set()
print(s)
print(type(s))
```

```python
# Its not a set its consider dictionary
s = {}
print(s)
print(type(s))
```

```python
## Adding elements a single elements usind add() function
s = {1,2,3,4,5}
print(s)
print()
s.add(6)
print(s)
```

```python
## Add multiple value in the form of list using update() function.......
s = {1,2,3,4,5}
print(s)
print()
s.update([6,7,8])
print(s)
```

```python
## Remove Elements into the set using remove()
s = {1,2,3,4,5}
print(s)
s.remove(3)
print(s)
```

```python
## Discard is also use for remove element from the sets.
## If element are not present then its give origunal set not gives error.
s = {1,2,3,4,5}
print(s)
print()
s.discard(7) ## 7 not present in sets
print(s)
print()
s.discard(4)
print(s)
```

```python
## Pop() also used for removing the elements.
s = {1,2,3,4,5}
print(s)
s.pop()
print()
```

```python
## Only one way for accessing elements in a sets.
s = {1,2,3,4,5}
```

```python
## Union Operation on  a sets
a = {1,2,3}
b = {3,4,5,6}
## First Method
c = a.union(b)
print(c)
print()

## Second Method
c = a | b
print(c)
```

```python
## Intersaction Operation on  a sets
a = {1,2,3}
b = {3,4,5,6}
## First Method
c = a.intersection(b)
print(c)
print()

## Second Method
c = a & b
print(c)
```

```python
## Difference Operation on  a sets
a = {1,2,3}
b = {3,4,5,6}
## First Method
c = a.difference(b)
print(c)
```

```python
## Symmetric Difference Operation on  a sets
## Remove common elements
a = {1,2,3}
b = {3,4,5,6}
## First Method
c = a.symmetric_difference(b)
print(c)
```

```python
## Check duplicate value.
l =[1,2,3,4,2,3]
unique = set()
dup = False
for i in l:
    if i in unique:
        dup = True
        break
    else:
        unique.add(i)
print(dup)
print(unique)
```

```python
l = [1,2,4,5]
n = 5
sum_seq = (n*(n+1))//2
print(sum_seq)
sum_l = 0
for i in l:
    sum_l += i
print(sum_l)
missing = sum_seq - sum_l
print(missing)
```
