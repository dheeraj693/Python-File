## Python Set Data Type – Practice Questions

#### Basic Questions (Foundation)

```python
# 1 Create a set from a list: Input [1, 2, 3, 2] → Output {1, 2, 3}
l = [1,2,3,2]
s = set(l)
print(s)
```

```python
#2 Add an element to a set: s = {1, 2, 3}, add 4 → {1, 2, 3, 4}
s = {1, 2, 3}
s.add(4)
print(s)
```

```python
# 3 Remove an element: s = {1, 2, 3}, remove 2 → {1, 3}
s = {1,2,3,4}
s.remove(2)
print(s)
```

```python
# 4 Check if an element exists in a set: s = {10, 20, 30}, check 20 → True
s =  {10,20,30}
for i in s:
    if i == 20:
        print("True")
```

```python
#5 Find the length of a set without using len()
s = {1,2,3,4,5,6}
count = 0
for i in s:
    count += 1
print(count)
```

```python
# 6 Convert a list to a set: [1, 2, 2, 3] → {1, 2, 3}
l = [1,2,3,4,5]
s = set(l)
print(s)
```

```python
# 7 Clear all elements from a set
s = {1,2,3,4,5}
print(s)
s.clear()
print("Clear sets : ",s)
```

```python
# 8 Copy a set
s = {1,2,3,4,5}
new = s.copy()
print(new)
```

```python
# 9 Iterate over a set and print all elements
s = {1,2,3,4,5,6,7}
for i in s:
    print(i)
```

```python
# 10 Create an empty set correctly (hint: {} creates a dictionary)
s = set()
print("Empty Sets :",s)
```

#### Basic Logic Building Questions

```python
# 1 Remove duplicates from a list: [1,2,2,3,4,4] → [1,2,3,4]
l = [1,2,2,3,4,4,5]
s = set(l)
print(s)
```

```python
#2 Find common elements in two lists: [1,2,3] & [2,3,4] → {2,3}
l_1 = [1,2,3]
l_2 = [2,3,4]
s1 = set(l_1)
s2 = set(l_2)
print(s1)
print(s2)
s3 = s1.intersection(s2)
print(s3)
```

```python
# 3 Find unique elements (present in only one list)
l_1 = [1,2,3]
l_2 = [2,3,4]
s1 = set(l_1)
s2 = set(l_2)
print(s1)
print(s2)
s3 = s1.difference(s2)
print(s3)
```

```python
# 4 Check if a list contains duplicates
l = [1,2,3,2,4,5,4]
if len(set(l)) < len(l):
    print("Duplicates are present : True")
else:
    print("Duplicates are not present : False")
```

```python
## Next Method for finding duplicates.
l = [1,2,3,2,4,5,4]
unique = []
duplicate = False
for i in l:
    if i in unique:
        duplicate = True
        break
    else:
        unique.append(i)
print("Contain Duplicate : ",duplicate)
```

```python
# 5 Find the missing number in a sequence: [1,2,4,5] → 3
l = [1,2,4,5]
n = 5
total_sum = (n*(n+1))//2
sum_l = 0
for i in l:
    sum_l += i
missing_num = total_sum - sum_l
print(missing_num)
    
```

```python
# 6 Check if two lists are equal (ignoring order)
l_1 = [1,2,3,4]
l_2 = [4,2,3,1]
if set(l_1) == set(l_2):
    print("List are equal")
else:
    print("List are not equal")
```

```python
# 7 Find the first duplicate element in a list
l = [1,2,3,4,3,2,5,6 ]
for i in range(len(l)+1):
    for j in range(0,len(l)-i-1):
        if l[j] > l[j+1]:
            l[j],l[j+1]=l[j+1],l[j]
new = l
print(new)
for i in range(len(new)):
    if new[i] == new[i+1]:
        print(new[i])
        break
        
```

```python
# Next Method using sets
l = [1,2,3,4,3,2,5,6 ]
unique = set()
for x in l:
    if x in unique:
        print("First Duplicate : ",x)
        break
    else:
        unique.add(x)
print(unique)
```
