## if - elif statement 

```python
# 1. Check if a number is positive, negative or zero using elif .
n = int(input("Enter the number : "))
if n > 0:
    print(f"{n} is positive number")
elif n < 0:
     print(f"{n} is negetive  number")
else:
     print("Zero")
```

```python
# 2.Take a number and print "Small" if it is less than 10, "Medium" if it is between 10 and 50, otherwise print "Large".

n = int(input("Enter the number : "))
if n < 10:
    print("It's a small number")
elif n >= 10 and  n <= 50:
    print("It's a medium number")
else:
    print("It's a large number")
```
