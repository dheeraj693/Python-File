```python
## Write a function to check number is prime of not.

def is_prime(num):
    
    for i in range(2,num):
        if num % i == 0:
            return False
    return True

is_prime(33)
            
```

```python
## Write a function to check the polindrome
def reverse(num):
    rev = 0
    temp = num

    while num > 0:
        digit = num%10
        rev = rev * 10 + digit
        num = num // 10
    if temp == rev:
        return True
    else:
        return False
reverse(121)
```

```python
## Write a function to return the reverse string.
def reverse_string(text):
    rev = ''
    for i in range(len(text)-1,-1,-1):
        rev += text[i]
    return rev
reverse_string("Dheeraj")
```

### Parameter and Arguments.

```python
## Type of Arguments.

# 1. Default Arguments.
# 2. Positional Arguments.
# 3. Keyword Arguments.
# 4. Variable length Argumnets. (*arg)
# 5. Variable length keyword arguments.(**kwarg)
```

```python
# 1. Default Arguments.
def greet(name='User'):
    print(f'Hello , {name} . May I help you.?')


greet()
```

```python
# 2. Positional Arguments.
def subtract(a,b):
    
    return a-b

subtract(8,2)
```

```python
def detail(name,age):
    print(f'My name in {name} . My age is {age} ')

detail('Dheeraj',22)
```

```python
def detail(name,age):
    print(f'My name in {name} . My age is {age} ')

detail(22,'Dheeraj')
```

```python
# 3. Keyword Arguments.
def detail(name,age):
    print(f'My name in {name} . My age is {age} ')

detail(age=22,name='Dheeraj')
```

```python
# 4. Variable length Argumnets. (*arg)
def sum_n_number(*arg):
    return sum(arg)

sum_n_number(2,3,4,5,6)
```

```python
# 5. Variable length keyword arguments.(**kwarg)
def details(**kwarg):
    return kwarg

details(name='Dheeraj',age = 22)
```

```python
## Call By value and call by referance.

# Call by value occure when data are immutable(int,float,string,tuple,etc.)
def change(a):
    a = a + 10
    print(a)

a = 10
change(a)
```

```python
# Call by referense occure when data are mutable(list,dictionary,sets).
def add_element(l):
    l.append(10)
    print(l)

l = [1,2,3]
add_element(l)
print(l)
```

```python
## Scope of Variable
# 1. Local Variable
# 2. Global Variable
```

```python
a = 10 # Global Variable

def display():
    b = 20 # Local Variable

    print(a)

display()
```

```python
## Change Global variable value using global keyword.

a = 10 # Global Variable

def display():
    global a # Used for change the global value of a.
    b = 20 # Local Variable
    a = 20
    print(b)

display()
print(a)
```

## High Order Function And Recurssion.


```python
## High Order Function
# High order the function  are the function in which another
# function is passed as an arguments.


def is_even(n): # First order function
    if n % 2==0:
        return True
    else:
        return False

def filter(func,li): # High order function
    result = []
    for i in li:
        if func(i):
            result.append(i)
    return result

li = [1,2,3,4,5,6,7,8]

## When 
filter(is_even,li)
```

```python
# MAp  
# Map is a high order function which takes first takes arguments as an function
# and second order argumnets which as an itrable..and it map all the elements
# of the iterable with that function and this return a mapped object.

# How to take multiple user input 

li = list(map(int , input().split()))
print(li)
```

```python
## Filter
# It is also a high order function which accepts
# another function as an argumnets but into this funnction if condition 
# is satifies with the element then will add that into the function
# or this function neglect that..
def is_even(n):
    if n%2==0:
        return True
    else:
        return False


li = [1,2,3,4,5,6,7]

list(filter(is_even , li))
```

```python
# Reduce 
#The reduce() function is used to apply a binary function cumulatively to
# all items in an itrable , condensing them into a single final value.

from functools import reduce

def add(a, b):
    return a + b

li = [1, 2, 3, 4, 5]
result = reduce(add, li)
print(result)   # Output: 15
```

```python
# Decorator
# Decoraters are the high oerdeer function which accepts the another 
# Function as an arguments and modify that function without changing 
# the its original function . It inhanced the functionality of the existing
#function without changing the original function code.

def add_func(func):
    def byee():
        print("Hiii...")
        func()
        print("Byeee...")
    return byee

@add_func

def say():
    print("Hello")
say()
    
```

```python
# Lamda Fucntion
# LAmbda function are the function which are define in one line andd this
# are also called as anonymus function.
# these are define with the lambda keyword
add = lambda x,y : x+y
add(3,4)
```
