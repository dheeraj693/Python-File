## 🟢 Easy Level

```python
# 1.Write a function to check if a number is even or odd.
def even_odd(num):
    if num % 2 == 0:
        return (f'{num} is Even Number')
    else:
        return (f"{num} is Odd Number")

n = int(input("Enter the number : "))
even_odd(n)
```

```python
# 2.Create a function that returns the square of a number.
def square(num):

    return (f"{num*num} is Square of {num}")

n = int(input("Enter the number : "))
square(n)  
```

```python
# 3.Write a function to find the maximum of two numbers.
def maximum(x,y):
    if x>y:
        return (f"Maximum is x : {x}")
    else:
        return (f"Maximum is y : {y}")
n = int(input("Enter the number x : "))
m = int(input("Enter the number y : "))
maximum(n,m)
```

```python
# 4.Create a function to count vowels in a string.
def count(text):
    
    vowels = 'aeiou'
    count_vowels = 0
    for i in text.lower():
        if i in vowels:
            count_vowels += 1
    return count_vowels



s = input("Enter the string : ")
count(s)
```

```python
# 5.Write a function that takes a list and returns the sum of elements.
def sum_elements(num):
    sum = 0
    while num>0:
        digit = num%10
        sum +=  digit
        num = num//10
        
    return sum


n = int(input("Enter the number x : "))
sum_elements(n)
```

## 🟡 Medium Level

```python
# 6. Write a function to check whether a number is prime.
def prime(num):
    

    for i in range(2,num):
        if num % i == 0:
            return False
    return True


n = int(input("Enter the number : "))
prime(n)
```

```python
# 7.Create a function to find factorial of a number.
def factorial(num):

    fact = 1
    if num == 0 and num == 1:
        return 1
    else:
        for i in range(1,num+1):
            fact = fact*i
        return fact
            
n = int(input("Enter the number : "))
factorial(n)
```

```python
# 8.Write a function to reverse a string.
def reverse(text):

    rev = ''
    for i in range(len(text)-1,-1,-1):
        rev += text[i]
    return rev


s = input("Enter the string : ")
reverse(s)
```

```python
# 9.Create a function that checks if a string is a palindrome.
def palindrom(text):
    rev = ''
    for i in range(len(text)-1,-1,-1):
        rev += text[i]

    if rev == text:
         return "Palindrome string"
    else:
         return "Not Palindrome"
        


s = input("Enter the string : ")
palindrom(s)
    
```

```python
# 10.Write a function to find the largest element in a list without using max().
def largest(l):

    
    large = 0
    for i in range(len(l)):
        if  l[i] > large:
            large = l[i]
    return large

        
## Taking user input in a list   
l = list(map(int, input("Enter numbers: ").split()))
print(l)

## Function call
largest(l)
    
```

```python
# Write a Function to print second largest number.
def second_largest(l):
    
    largest = 0
    second_large = 0
    for i in range(len(l)):
        if largest < l[i]:
            second_large = largest
            largest = l[i]
        elif l[i]>largest and l[i] > second_largest:
            second_large = l[i]
    return second_large



l = list(map(int, input("Enter numbers: ").split()))
print(l)
second_largest(l)
```
