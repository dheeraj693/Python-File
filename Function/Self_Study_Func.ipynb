## Type Of Aguments.
#### 1. Default Argument
#### 2. Positional Argumnet
#### 3. Keyword Argument

```python
## 1. Default Arguments.
## Set A value for taking as default
def power(a=1,b=1):
    """
    This Function return the calculation  power of number 
    input - a,b int
    output - a = 2,b=3 return a**b = 8
    """
    return a**b
# Calling Function

power(4) ## Take value of b as default......
    
```

```python
## 2. Positional Arguments
## Fixed the position/Order of variable when creating a function

def power(a=1,b=1):
    """
    This Function return the calculation  power of number 
    input - a,b int
    output - a = 2,b=3 return a**b = 8
    """
    return a**b



power(4,2)
```

```python
## 3. Keyword Arguments
# Keyword arguments me position se koi lena dena nhi hota h
# Ye depend karta h kis variable/keywords ko kaun sa value de rhe h

def power(a=1,b=1):
    """
    This Function return the calculation  power of number 
    input - a,b int
    output - a = 2,b=3 return a**b = 8
    """
    return a**b


## Used keywords when callig function
power(b=4,a=2)
```

## Importants : *args and **kwargs

```python
## *args :-
# Allow us to pass a variable number of non-keywords arguments to a function
# Used for given multiple input at a time
def multiply(*args):

    product = 1
    for i in args:
        product = product *i
    print(args)
    return product

# Multiple input  given in function
multiply(2,3,3)
```

```python
def multiply(*args):

    product = 1
    for i in args:
        product = product *i
    print(args)
    return product

# Multiple input  given in function
a = int(input())
b = int(input())
c =  int(input())
multiply(a,b,c)
```

```python
## **kwargs :- keywords arguments
# Its allow us to pass any number of keyword arguments.
# Keywords arguments means that they contain a key-value pairs ,  
# Like a python dictionary

def display(**kwargs):

    for key,value in kwargs.items():
        print(key , '->',value)


display(name = 'Dheeraj',title='Prajapati',age=22)
        
```

```python
##
def num(y): # Y is a local variable.
    print(y)
    print(y+1)

x = 5  # x is a global variable.
num(x)
```
