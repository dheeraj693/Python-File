```python
# Q3. Sorting Using Lambda
# You are given a list of tuples where each tuple contains a student's name and age:

# students = [("Aman", 25), ("Riya", 20), ("Karan", 23)]

# Sort the list in ascending order based on age using a lambda function.
# Output:
# [("Riya", 20), ("Karan", 23), ("Aman", 25)]

students = [("Aman", 25), ("Riya", 20), ("Karan", 23)]
result = sorted(students , key=lambda x:x[1])
print(result)
```

```python
# Q5. Filter Even Numbers

# Given a list of integers, use the filter() 
# function to extract all even numbers.

# nums = [1,2,3,4,5,6,7,8]

# Output: [2,4,6,8]

def even(n):
    if n % 2== 0:
        return True
    else:
        return False
def filter(func,nums):
    result = []
    for i in nums:
        if func(i):
             result.append(i)
    return result



nums = [1,2,3,4,5,6,7,8]
filter(even , nums)
```

```python
# Find Odd in a list
def odd(n):
    if n % 2 != 0:
        return True
    else:
        return False
def filter(func,li):
    result  = []
    for i in li:
        if func(i):
            result.append(i)
    return result

li = [1,2,3,4,5,6,7,8,9]
filter(odd,li)
        
```

```python
# Q6. Filter Long Words
# Given a list of strings, return only those words whose
# length is greater than 5.

# words = ["apple", "banana", "kiwi", "grapefruit"]

# Output: ["banana", "grapefruit"]
def large_len(word):
    if len(word) > 5:
        return True
    else:
        return False
def filter(func , words):
    result = []
    for i in words:
        if func(i):
            result.append(i)
    return result
words = ["apple", "banana", "kiwi", "grapefruit"]
filter(large_len , words)
```

```python
# Q7. Filter Prime Numbers

# Write a program to filter all prime numbers from a given 
#list using filter() .
# Input: [2,3,4,5,6,7,8,9]
# Output: [2,3,5,7]
def is_prime(n):
    if n <= 1:
        return False

    for i in range(2,n):
        if n%i == 0:
            return False
    return True
        
def filter(func , li):
    result = []
    for i in li:
        if func(i):
            result.append(i)
    return result

li = [2,3,4,5,6,7,8,9]
filter(is_prime , li)
```

```python
## Another way
def is_prime(n):
    if n <= 1:
        return False

    for i in range(2,n):
        if n%i == 0:
            return False
    return True
        

li = [2,3,4,5,6,7,8,9]
list(filter(is_prime , li))
```

```python
# Q8. Remove Empty Strings
# Given a list containing strings, remove all empty strings using filter().

# data = ["hello", "", "world", "", "python"]

# Output: ["hello", "world", "python"]
def remove(n):
    for i in n:
        word = n
        for j in word:
            if len(j) == 0:
                return False
        return True
data = ["hello", "", "world", "", "python"]
list(filter(remove , data))
```

```python
# Q12. Find Length of Each Word
# Given a list of words, use map() to create a list containing 
#the length of each word.
# Input: ["apple", "banana", "kiwi"]

# Output: [5,6,4]

def length(li):

    count = 0
    for i in li:
        word = i
        if i in word:
            count += 1
    return count


li =  ["apple", "banana", "kiwi"]
list(map(length , li))
```

## 🔷Section 4: Reduce Function

```python
# Q13. Sum of Elements

# Use reduce() to calculate the sum of all elements in a list.
# Input: [1,2,3,4]
# Output: 10

from functools import reduce

def add_num(a,b):
    return a+b

li = [1,2,3,4]
result = reduce(add_num , li)
print(result)
```

```python
# Find the product (multiplication) of all elements using reduce() .
# Input: [1,2,3,4]
# Output: 24

from functools import reduce

def multiply(a,b):
    return a*b

li = [1,2,3,4]
result = reduce(multiply , li)
print(result)
```

```python
# Q15. Maximum Element

# Use reduce() to find the maximum number in a list.
# Input: [10, 25, 5, 40]
# Output: 40

from functools import reduce

def max_num(a,b):
    if a>b:
        return a
    else:
        return b

li = [10,25,5,40]
result = reduce(max_num , li)
print(result)
```
