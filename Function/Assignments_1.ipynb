```python
# Write a recursive function:
# To calculate factorial
# Must use return , not print

def factorial(n):

    if n == 0 or n == 1:
        return 1
    else:
        return n*factorial(n-1)

x = int(input("Enter the number : "))
print(factorial(x))
```

```python
# 🔹16. Real Scenario Function
# Create a function:
# Accepts:
# name
# marks (any number using *args)
# Returns:
# average marks
# pass/fail (avg > 40)

def detail(name,*marks):

    avg = sum(marks)//len(marks)
    if avg > 40 :
        result = "Pass"
    else:
        result = "Fail"
    return avg,result

print(detail('Dheeraj',60,40,60))
print(detail('Dheeraj',70,80,50))


```
